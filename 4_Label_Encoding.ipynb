```python
import pandas as pd
```

```python
df = pd.DataFrame({"name":['wscube', 'cow', 'cat', 'dog', 'black', 'dipesh']})
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df["en_name"] = le.fit_transform(df["name"])
```
