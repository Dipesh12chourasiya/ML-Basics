```python
import pandas as pd
df = pd.DataFrame({"Size":['xs','s','m','l','xl','xxl','xs','s','m','l','xl','xs','s','m','l','xl','xxl','m']})
df.head(3)
```

```python
ord_data = [['xs','s','m','l','xl','xxl']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = ord_data)
oe.fit(df[['Size']])
```

```python
df["Size_En"] = oe.transform(df[['Size']])
```
